```python
# Define if-statement

def price_label(row):

  if row['price'] <= 5:
    return 'Low'
  elif (row['price'] > 5) and (row['price'] <= 15):
    return 'Medium'
  elif row['price'] > 15:
    return 'High'
  else: return 'No data'
```

```python
# Use if-statement. Create new column price_range and fill it up with if-statement's results
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# Check values in new column price_range
df['price_range'].value_counts(dropna = False)
```

```python
# STEP 2 Create new column where the two busiest days are labelled as 'Busiest days'. Also, label the two slowest days
# Create for-loop to perform task

result_1 = []
for value in df_ords_prods_merged['order_day_of_week']:
    if value == 0 or value == 1:
        result_1.append('Busiest days')
    elif value == 3 or value == 4:
        result_1.append('Least busy')
    else:
        result_1.append('Regularly busy')
```

```python
# STEP 4 Identify busiest hours of day
# Create new column busiest_period_of_day containing labels Most orders, Average orders, and Fewest orders
```

```python
# Find frequency of order_hour_of_day
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# STEP 4 CONTINUED
# Having found the frequency, divide the hours of day in the three categories requested (24/3=8)
# Create for-loop to assign labels about how busy days of week are

result_2 = []
for value in df_ords_prods_merged['order_hour_of_day']:
    if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9:
        result_2.append('Most orders')
    elif value == 17 or value == 8 or value == 18 or value == 19 or value == 20 or value == 7 or value == 21 or value == 22:
        result_2.append('Average orders')
    else:
        result_2.append('Fewest orders')
```
